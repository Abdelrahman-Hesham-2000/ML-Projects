## Importing Necessary libraries

## Preprocessing

```python
dt.head(10)
```

#### Feature Extraction

```python
dt['Datetime'] = pd.to_datetime(dt['Datetime'], format='%m/%d/%Y %H:%M')
```

```python
dt['Month'] = dt['Datetime'].dt.month
```

```python
dt['TimeFloat'] = dt['Datetime'].dt.hour + dt['Datetime'].dt.minute / 60
```

```python
dt = dt.drop(columns=['Datetime'], axis= 1)
```

```python
dt.head()
```

## Modeling

```python
X = dt.drop(columns=['PowerConsumption_Zone1','PowerConsumption_Zone2','PowerConsumption_Zone3'], axis= 1)
y1 = dt['PowerConsumption_Zone1']
y2 = dt['PowerConsumption_Zone2']
y3 = dt['PowerConsumption_Zone3']
```

#### Polynomial Rregression

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline

degree = 5

X_train, X_test, y1_train, y1_test = train_test_split(X, y1, test_size=0.2, random_state=42)
_, _, y2_train, y2_test = train_test_split(X, y2, test_size=0.2, random_state=42)
_, _, y3_train, y3_test = train_test_split(X, y3, test_size=0.2, random_state=42)

def poly_regression(X_train, X_test, y_train, y_test, degree=2):
    model = make_pipeline(PolynomialFeatures(degree), LinearRegression())
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = root_mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return rmse, r2

rmse1, r2_1 = poly_regression(X_train, X_test, y1_train, y1_test, degree)
print(f"Zone1 - RMSE: {rmse1:.2f}, R2: {r2_1:.2f}")

rmse2, r2_2 = poly_regression(X_train, X_test, y2_train, y2_test, degree)
print(f"Zone2 - RMSE: {rmse2:.2f}, R2: {r2_2:.2f}")

rmse3, r2_3 = poly_regression(X_train, X_test, y3_train, y3_test, degree)
print(f"Zone3 - RMSE: {rmse3:.2f}, R2: {r2_3:.2f}")
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestRegressor

def rf_regression(X_train, X_test, y_train, y_test):
    model = RandomForestRegressor(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    rmse = root_mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return rmse, r2

rf_rmse1, rf_r2_1 = rf_regression(X_train, X_test, y1_train, y1_test)
print(f"Random Forest Zone1 - RMSE: {rf_rmse1:.2f}, R2: {rf_r2_1:.2f}")

rf_rmse2, rf_r2_2 = rf_regression(X_train, X_test, y2_train, y2_test)
print(f"Random Forest Zone2 - RMSE: {rf_rmse2:.2f}, R2: {rf_r2_2:.2f}")

rf_rmse3, rf_r2_3 = rf_regression(X_train, X_test, y3_train, y3_test)
print(f"Random Forest Zone3 - RMSE: {rf_rmse3:.2f}, R2: {rf_r2_3:.2f}")
```
