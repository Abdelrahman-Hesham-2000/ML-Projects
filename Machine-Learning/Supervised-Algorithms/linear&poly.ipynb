```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn.linear_model import LinearRegression
```

```python
# Sample data: predicting house price based on size
data = {'Size': [500, 750, 1000, 1250, 1500],
        'Price': [150000, 200000, 250000, 300000, 350000]}
df = pd.DataFrame(data)
df
```

```python
x=df[['Size']] # indep   # feature 
y=df['Price'] # dep  # target 
```

```python
model=LinearRegression()
model.fit(x,y)
```

```python
predictions=model.predict(x)
```

```python
plt.scatter(x,y,color='Blue',label='real data')
plt.plot(x,predictions,color='Red',label='predicted')
plt.xlabel("size ")
plt.ylabel("price ")
plt.legend()
plt.show()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
data = {'Time': [1, 2, 3, 4, 5],
        'Progression': [1, 4, 9, 16, 25]}
df = pd.DataFrame(data)
```

```python
X=df[['Time']]
y=df['Progression']
```

```python
# Transform features to polynomial features (degree 2)
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
# Create and train the model
model = LinearRegression()
model.fit(X_poly, y)
```

```python
predictions = model.predict(X_poly)
```

```python
# Plot the results
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, predictions, color='red', label='Polynomial Regression Line')
plt.xlabel('Time')
plt.ylabel('Progression')
plt.title('Polynomial Regression: Disease Progression Prediction')
plt.legend()
plt.show()
```

cost function

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
import numpy as np
```

```python
mse2=mean_squared_error(y,predictions)
print(mse2)
```

```python
mae=mean_absolute_error(y,predictions)
print(mae)
```

```python
rmse=np.sqrt(mse2)
print(rmse)
```
