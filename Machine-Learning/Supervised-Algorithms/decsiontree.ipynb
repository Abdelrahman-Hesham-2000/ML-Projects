```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import metrics 
import pandas as pd
```

```python
iris=load_iris()
df=pd.DataFrame(data=iris.data,columns=iris.feature_names)
df['target']=iris.target
df
```

```python
print(iris.keys())
```

```python
X=iris.data
y=iris.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
clf=DecisionTreeClassifier(criterion='gini')
```

```python
clf=clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
accuracy=metrics.accuracy_score(y_test,y_pred)
print(accuracy)
```

```python
import matplotlib.pyplot as plt 
from sklearn.tree import plot_tree
plt.figure(figsize=(10,5))
plot_tree(clf,filled=True,feature_names=iris.feature_names,class_names=iris.target_names,rounded=True)
plt.show()
```

```python
en=DecisionTreeClassifier(criterion='entropy',max_depth=3)
en.fit(X,y)

plt.figure(figsize=(10,5))
plot_tree(en,filled=True,feature_names=iris.feature_names,class_names=iris.target_names,rounded=True)
plt.show()
```

```python
from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt 
```

```python
data=fetch_california_housing()
```

```python
print(data.keys())
```

```python
print(data.DESCR)
```

```python
df=pd.DataFrame(data.data,columns=data.feature_names)
X=data.data
y=data.target
df
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
regressor=DecisionTreeRegressor(criterion='squared_error',max_depth=4)
regressor.fit(X_train,y_train)
```

```python
y_pred=regressor.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print(mse)
```

```python
features=data.feature_names
```

```python
plt.figure(figsize=(20,10))
plot_tree(regressor,filled=True,feature_names=features,class_names=data.target_names,rounded=True)
plt.show()
```
